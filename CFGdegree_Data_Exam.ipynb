```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset and return the resulting data frame

# Calculate the average duration of the movies in the dataset
average_duration = pd.to_numeric(netflix_df[netflix_df['type'] == 'Movie']['duration'].str.replace(' min', ''), errors='coerce').mean()

# Check if average_duration is NaN and replace with a default value if necessary
if pd.isna(average_duration):
    average_duration = 90  # default to 90 minutes if average calculation fails

# Fill missing duration values with the average duration for movies
for index, row in netflix_df.iterrows():
    if row['type'] == 'Movie' and pd.isna(row['duration']):
        netflix_df.at[index, 'duration'] = f'{int(average_duration)} min'

# Display the updated DataFrame
netflix_df.head(20)
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
# Check for duplicates
duplicates = netflix_df.duplicated()

# Print out the duplicate rows (if any)
if duplicates.any():
    print("Duplicate rows found:")
    print(netflix_df[duplicates])
else:
    print("No duplicate rows found.")

# Remove duplicates
netflix_df = netflix_df.drop_duplicates()

# Display the resulting DataFrame
print("DataFrame after removing duplicates:")
print(netflix_df)
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
netflix_df.columns = netflix_df.columns.str.title()
print("\nColumns renamed to start with a capital letter:")
print(netflix_df.columns)
print(netflix_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.

cleaned_netflix_df = pd.read_csv("cleaned_netflix_titles.csv")

# Filter the dataset for movies only
cleaned_movies_df = cleaned_netflix_df[cleaned_netflix_df['type'] == 'Movie']

# Plot the frequency of movies by year of release
plt.figure(figsize=(14, 7))
release_year_counts = cleaned_movies_df['release_year'].value_counts().sort_index()
release_year_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Frequency of Movies by Year of Release (Cleaned Data)', fontsize=16)
plt.xlabel('Year of Release', fontsize=16)
plt.ylabel('Number of Movies', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle=':', alpha=0.9)
plt.tight_layout()
plt.show()

```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.


# Filter the dataset for movies only
cleaned_movies_df = cleaned_netflix_df[cleaned_netflix_df['type'] == 'Movie']

# Split the 'listed_in' column into individual genres
genres = cleaned_movies_df['listed_in'].str.split(', ', expand=True).stack()

# Get the top 5 most popular genres
top_genres = genres.value_counts().nlargest(5)

# Plot a pie chart of the top 5 most popular genres
plt.figure(figsize=(20, 7))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("deep"), wedgeprops={'edgecolor': 'black'})
plt.title('Top 5 Most Popular Genres of Movies', fontsize=16, fontweight='bold', color='pink')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# 1 (2 marks). Running the code in this cell throws an error. Why do we get this error
  # and how do we fix it?
import numpy as np
# use this numpy array for all exercises in this section
arr = np.array([0, 56, 82, 500, 140, 43, 92, 10, 25, 11, 18, 26, 150, 240, 364, 196])
arr
```

```python
# 2 (5 marks). How many odd numbers do we have in the array? Show how we get this result using numpy

odd_numbers = arr % 2 == 1

# Extract the odd numbers from the original array
odd_number_list = arr[odd_numbers]

# Sum the boolean array to get the number of odd numbers
num_odd_numbers = np.sum(odd_numbers)

# Display the list of odd numbers and their count
print("List of odd numbers:", odd_number_list)
print("Number of odd numbers:", num_odd_numbers)
```

```python
# 3 (5 marks). Replace all odd numbers in arr with 999
# Replace all odd numbers with 999
arr[odd_numbers] = 999

# Display the updated array
print("Updated array with odd numbers replaced by 999:")
print(arr)

```

```python
# 4 (8 marks). The above is a 1D array. Convert it into a 4D array and reverse its columns.

# Desired output
# > array([[500,  82,  56,   0],
# >        [ 10,  92, 999, 140],
# >        [ 26,  18, 999, 999],
# >        [196, 364, 240, 150]])

# Original 1D array
arr = np.array([0, 56, 82, 500, 140, 999, 92, 10, 999, 999, 18, 26, 150, 240, 364, 196])

# Reshape the 1D array into a 4x4 2D array
reshaped_arr = arr.reshape((4, 4))

# Step 2: Reverse the columns of the reshaped array
reversed_columns_arr = reshaped_arr[:, ::-1]

# Display the result
print(reversed_columns_arr)

```
